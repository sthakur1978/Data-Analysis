# Matplotlib



```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
# Funtional Method
# basic method
plt.plot(x,y)
plt.show()
```

```python
# adding labels and title

plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.show()
```

```python
#Adding Multiplots

#plot 1
plt.subplot(1,2,1)
plt.plot(x,y,'r')
# plot 2
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object Oriented Method
# create figure objects and call methods out it

fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8]) # numbers need to be between 0 and 1

axes.plot(x,y)

axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
# plot multi plots


fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])

axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) # these are %, left, bottom, width, height

axes1.plot(x,y)
axes2.plot(y,x)

axes1.set_title('Larger Plot')
axes2.set_title('Smaller Plot')
```

```python
# Create subplots using OOP

fig,axes = plt.subplots(nrows =1 , ncols= 2)

for current_axe in axes:
    current_axe.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows =1 , ncols= 2)
axes[0].plot(x,y)
axes[1].plot(x,y)

# using indexing
```

```python
# Fig size and DPI

fig = plt.figure(figsize = (8,2))

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows = 2, ncols = 1,figsize = (8,2))

axes[0].plot(x,y)
axes[1].plot(x,y)
```

```python
# Save figure

fig.savefig('mypic.jpg')
```

```python
# Legend

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label= 'X Square')
ax.plot(x,x**3, label = 'X Cube')

ax.set_title('Title',)
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.legend()
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

#ax.plot(x,y, color= 'Red') # adding colors

#ax.plot(x,y, color= '#FF8C00') # Hex codes

# edit line width
#ax.plot(x,y, color= 'Red', linewidth = 5) 

# using alpha for tranperancy

#ax.plot(x,y, color= 'Red', linewidth = 5, alpha = 0.5) 

# Linestyle

#ax.plot(x,y, color= 'Red', lw = 5, alpha = 0.5, ls = '-.') 

# using Markers

ax.plot(x,y, color= 'Red', lw = 3, alpha = 0.5, ls = '--', Marker = 'o', markersize = 20, markerfacecolor= 'yellow',
       markeredgewidth = 3, markeredgecolor = 'green') 
```

```python
# Plot Range
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color= 'Red', lw = 3, alpha = 0.5, ls = '--')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```

## Using Seaborn

```python
## Distribution Plots

import seaborn as sns
%matplotlib inline
```

```python
# use tips data set

tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# show univarite distribution

#sns.distplot(tips['total_bill'])

sns.distplot(tips['total_bill'], kde = False, bins = 30)
```

```python
# Joint plot for bivariate data

sns.jointplot(data = tips, x= 'total_bill', y ='tip', kind = 'kde')
```

```python
# using Pairplots

sns.pairplot(tips, hue = 'sex')
```

```python
# Rug plot

sns.rugplot(tips['total_bill'])
```

```python
# Plotting Categorical Data

sns.barplot(data = tips, x = 'sex',y ='total_bill')
```

```python
import numpy as np
sns.barplot(data = tips, x = 'sex',y ='total_bill', estimator = np.std ) # default is mean
```

```python
sns.countplot(data = tips, x = 'sex') # count of gender, only x is passed
```

```python
# Box Plot shows distribution of categorical data

sns.boxplot(data = tips,x='day',y ='total_bill')

# adding Hue for another category

#sns.boxplot(data = tips,x='day',y ='total_bill', hue ='smoker')
```

```python
#violin Plot

#sns.violinplot(data = tips,x='day',y ='total_bill')

#using another category

sns.violinplot(data = tips,x='day',y ='total_bill', hue = 'sex')
```

```python
#sns.stripplot(data = tips,x='day',y ='total_bill', jitter = True)

# add another category using Hue

sns.stripplot(data = tips,x='day',y ='total_bill', jitter = True, hue = 'sex',dodge = True)
```

```python

sns.swarmplot(data = tips,x='day',y ='total_bill')
```

```python
sns.violinplot(data = tips,x='day',y ='total_bill' )
sns.swarmplot(data = tips,x='day',y ='total_bill',color = 'black')
```

```python
sns.factorplot(data = tips,x='day',y ='total_bill', kind = 'bar') # use factorplot and use the kind to define the plot
```
